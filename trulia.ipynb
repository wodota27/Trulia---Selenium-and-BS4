```python
import time
from bs4 import BeautifulSoup
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.chrome.service import Service
import requests
import pandas as pd
import urllib.parse 
```

```python

website = 'https://www.trulia.com/CA/San_Diego/' 
```

```python
website
```

```python
options = Options()
options = webdriver.ChromeOptions()
options.add_argument('--headless')
options.add_argument('--disable-gpu')
#driver = webdriver.Chrome(chrome_options=options)
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))
driver.get(website)
time.sleep(3)
page = driver.page_source
driver.quit()
```

```python
soup = BeautifulSoup(page, 'html.parser')
```

```python
soup
```

```python
container = soup.find_all('li', attrs={'class':'SearchResultsList__WideCell-sc-14hv67h-2'})
```

```python
len(container)
```

```python
results_update = []
```

```python
for r in container:
    if r.has_attr('data-testid'): 
        results_update.append(r)
```

```python
len(results_update)
```

```python
url_part_1 = 'https://www.trulia.com'
```

```python
url_part_2 = []

for item in results_update:
    
    for link in item.find_all('div', {'data-testid':'property-card-details'}):
        url_part_2.append(link.find('a').get('href'))
```

```python
len(url_part_2)
```

```python
url_joined = []

for link_2 in url_part_2:
    url_joined.append(urllib.parse.urljoin(url_part_1,link_2))
```

```python
address = []
bedrooms = []
price = []


for link in url_joined:
    options = Options()
    options = webdriver.ChromeOptions()
    options.add_argument('--headless')
    options.add_argument('--disable-gpu')
    #driver = webdriver.Chrome(chrome_options=options)
    driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))
    driver.get(link)
    time.sleep(1)
    page = driver.page_source
    driver.quit()
    
    soup = BeautifulSoup(page, 'html.parser')
    
    try:
        address.append(soup.find('span', {'data-testid':'home-details-summary-headline'}).get_text())
    except:
        address.append('')
        
    # bedrooms 
    try:
        bedrooms.append(soup.find('li', {'data-testid':'bed'}).get_text())
    except:
        bedrooms.append('')
        
    try:
        price.append(soup.find('h3', {'data-testid':'on-market-price-details'}).get_text())
    except:
        price.append('')
        
     
    output = {'Address': address, 'Bedrooms':bedrooms, 'Price':price}
```

```python
#multiple pages
address = []
bedrooms = []
price = []


url_part_1 = 'https://www.trulia.com'

for i in range(1,4):
    
    website = 'https://www.trulia.com/CA/San_Diego/' +str(i) +'_p/'
    
    options = Options()
    options = webdriver.ChromeOptions()
    options.add_argument('--headless')
    options.add_argument('--disable-gpu')
    #driver = webdriver.Chrome(chrome_options=options)
    driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))
    driver.get(website)
    time.sleep(3)
    page = driver.page_source
    driver.quit()
    
    soup = BeautifulSoup(page, 'html.parser')
    
    container = soup.find_all('li', attrs={'class':'SearchResultsList__WideCell-sc-14hv67h-2'})

    results_update = []
    
    # only results with attribute "data-testid"
    for r in container:
        if r.has_attr('data-testid'): 
            results_update.append(r)
            
    relative_url = []
    
    # loop thorugh results
    for item in results_update:
    
        for link in item.find_all('div', {'data-testid':'property-card-details'}):
            relative_url.append(link.find('a').get('href'))
    
    # empty list (url joined)
    url_joined = []

    for link_2 in relative_url:
        url_joined.append(urllib.parse.urljoin(url_part_1,link_2))
        
    for link in url_joined:
        options = Options()
        options = webdriver.ChromeOptions()
        options.add_argument('--headless')
        options.add_argument('--disable-gpu')
        #driver = webdriver.Chrome(chrome_options=options)
        driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))
        driver.get(link)
        time.sleep(1)
        page = driver.page_source
        driver.quit()
        
        soup = BeautifulSoup(page, 'html.parser')
        
        try:
            address.append(soup.find('span', {'data-testid':'home-details-summary-headline'}).get_text())
        except:
            address.append('')

        # bedrooms 
        try:
            bedrooms.append(soup.find('li', {'data-testid':'bed'}).get_text())
        except:
            bedrooms.append('')

        try:
            price.append(soup.find('h3', {'data-testid':'on-market-price-details'}).get_text())
        except:
            price.append('')


        output = {'Address': address, 'Bedrooms':bedrooms, 'Price':price}
        

    
    
    

```
